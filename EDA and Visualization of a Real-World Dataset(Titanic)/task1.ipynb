```python
def handle_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Capping the outliers
    df[column] = np.where(df[column] < lower_bound, Q1, df[column])
    df[column] = np.where(df[column] > upper_bound, Q3, df[column])

#Apply the function to the relevant columns
outlier_columns = ['age', 'fare', 'sibsp', 'parch']
for col in outlier_columns:
    handle_outliers(col)

#Log transformation for 'Fare' to reduce skewness
df['fare'] = np.log1p(df['fare'])  # log1p handles log(0) safely

#print(df[outlier_columns].describe())
```
