```python
#filling missing value with mean of Age: 
mean=df['Age'].mean()
df['Age'].fillna( mean , inplace=True)

#filling missing value with mode of Sex:
df['Sex'].replace('Male'==1, 'Female'==0)
mode = df['Sex'].mode()[0]  # Calculate the mode of the 'Sex' column
df['Sex'].fillna(mode, inplace=True)

#Changing the DType of Age:
df['Age'] = df['Age'].astype(object)
df['FraudFound_P'] = df['FraudFound_P'].astype(object)
```

```python
# Drop the 'Witness' column from the DataFrame
df.drop(columns=['WitnessPresent'], inplace=True)


# Fill missing values in the 'MaritalStatus' column with the mode
mode_marital_status = df['MaritalStatus'].mode()[0]
df['MaritalStatus'].fillna(mode_marital_status, inplace=True)
```

```python
def fraud_detection_interface():
    print("\n" + "="*50)
    print("🔍 FRAUD-DETECTION SYSTEM TEST INTERFACE")
    print("="*50)
    
    try:
        # Collect key features from the user
        print("\nEnter claim details:")
        age = float(input("• Policy Holder Age: "))
        days_policy_accident = int(input("• Days Since Policy Start: "))
        past_claims = int(input("• Past Number of Claims: "))
        num_supplements = int(input("• Number of Supplements: "))
        police_report = input("• Police Report Filed? (yes/no): ").lower().strip()
        accident_area = input("• Accident Area (urban/rural): ").lower().strip()
        vehicle_price = input("• Vehicle Price (high/medium/low): ").lower().strip()
        
        # Create a dictionary with default values for all features in x_train.
        input_dict = {}
        for col in x_train.columns:
            input_dict[col] = 0  
        
        input_dict['Age'] = age
        input_dict['Days_Policy_Accident'] = days_policy_accident
        input_dict['PastNumberOfClaims'] = past_claims
        input_dict['NumberOfSuppliments'] = num_supplements
        
    
        input_dict['PoliceReportFiled_yes'] = 1 if police_report == 'yes' else 0
        input_dict['PoliceReportFiled_no'] = 1 if police_report == 'no' else 0
        
        input_dict['AccidentArea'] = 1 if accident_area == 'urban' else 0
        
        input_dict['VehiclePrice_high'] = 1 if vehicle_price == 'high' else 0
        input_dict['VehiclePrice_medium'] = 1 if vehicle_price == 'medium' else 0
        input_dict['VehiclePrice_low'] = 1 if vehicle_price == 'low' else 0
        
        test_input_df = pd.DataFrame([input_dict])
        test_input_df = test_input_df[x_train.columns]
        
        prob_fraud = model.predict_proba(test_input_df)[0][1]
        prediction = "FRAUD" if prob_fraud > 0.5 else "LEGITIMATE"
        
        print("\n" + "="*50)
        print("PREDICTION RESULT:")
        print("Status:", prediction)
        print("Confidence (fraud probability):", f"{prob_fraud:.1%}")
        print("="*50)
        
    except ValueError:
        print("\nError: Please enter valid numbers where required")
    except Exception as e:
        print(f"\n An error occurred: {str(e)}")

fraud_detection_interface()
```
